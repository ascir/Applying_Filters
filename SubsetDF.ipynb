# Functionality
This module enables the user to apply various filters to a dataset and perform operations such as 'and' and 'on' to combine these subsets.

##  Required parameters
* Input dataframe  


## Default parameters
*  The location to which the outPut csv file is saved

## Input parameters

### Following inputs are to be entered in the form of a python dictionary:

* ColumnName: Name of the column over which data is subset 
* Condition: 
  * for numeric: gt(Greater than, lt(Lesser than), et(Equal to), gte(Greater than and equal to), lte(Lesser than and equal     to), ne(Not equal to), between(filter between two values)
  * for categorical: in(Filter those rows which lie in a certain category), not in(Filter out those rows which do not lie       in a category)
* Value: A value in the colName specified according to which dataframe is subset
* Name_df: Name of the new dataframe created after subsetting
* Operation: operator/ Name of the operation to be performed: 'and'/'or
   * AND . Both relations must be true for the complex expression to be true.
   * OR . If either relation is true, the complex expression is true.
   * Refer to Test 1 for an example

### Sample: 
  #### alcohol = {'ColumnName': ['beer_servings', 'continent', 'wine_servings'],
  #### 'Condition': [ 'between', 'in', 'gt'],
  #### 'Value':[[100, 200], 'Asia', 100],
  #### 'Name_df':['new_1', 'new_2', 'new_3'],
  #### 'Operation':['and', 'or']}
    **Please note number of operations that can be performed is one less than the number of filters applied

### Following inputs are to be entered during function call:


* df: Dataframe to be subsetted 
* Name of the dictionary that was entered

## Return
*  Appropriately filtered subset of the input dataframe

# Code

```python
import pandas as pd
import numpy as np
from collections import OrderedDict
df1 = pd.read_csv('http://bit.ly/imdbratings')
df2 = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
def applfilter(df, sbset):
    mov= pd.DataFrame.from_dict(sbset, orient='index')#converting the dict to df
    movs= mov.transpose()
    new_col = movs.columns.values #getting column values of df
    new_col[4] = 'Operation' 
    movs.columns= new_col #Changing value of 4th column to 'Operation'
    movs.Operation= movs.Operation.shift(1) 
    movs

    for i in range(0, len(movs.index)):
        #performing numerical and categorical filtering
        if(movs.loc[i][1] == 'gt'):
            movs.loc[i][3] = df[movs.loc[i][0]] > movs.loc[i][2]
        elif(movs.loc[i][1] == 'lt'):
            movs.loc[i][3] = df[movs.loc[i][0]] < movs.loc[i][2]          
        elif(movs.loc[i][1] == 'et'):
            movs.loc[i][3] = df[movs.loc[i][0]] == movs.loc[i][2]       
        elif(movs.loc[i][1] == 'gte'):
            movs.loc[i][3] = df[movs.loc[i][0]] >= movs.loc[i][2]
        elif(movs.loc[i][1] == 'lte'):
            movs.loc[i][3] = df[movs.loc[i][0]] <= movs.loc[i][2]
        elif(movs.loc[i][1] == 'ne'):
            movs.loc[i][3] = df[movs.loc[i][0]] != movs.loc[i][2]
        elif(movs.loc[i][1] == 'between'):
            movs.loc[i][3] = (df[movs.loc[i][0]] > movs.loc[i][2][0]) & (df[movs.loc[i][0]] < movs.loc[i][2][1])
        elif(movs.loc[i][1] == 'in'):
            movs.loc[i][3] = df[movs.loc[i][0]] == movs.loc[i][2]
        elif(movs.loc[i][1] == 'not in'):
            movs.loc[i][3] = df[movs.loc[i][0]] != movs.loc[i][2]
        else:
            print('Error: Invalid Input')
            
    for x in range(1, len(movs.index)):
        #performing operations on filtered datasets
        if movs.loc[x][4] == 'and':
            boolean = (movs.loc[x-1][3])&(movs.loc[x][3])
            if(len(movs.index) > 1): #checking if more than 1 filter applied
                for y in range(2, len(movs.index)):
                    if(movs.loc[y][4] == 'and'):
                        boolean = (boolean) & (movs.loc[y][3])#applying filter to several dataset in loop
                    elif(movs.loc[y][4] == 'or'):
                        boolean = (boolean) | (movs.loc[y][3])
                    else:
                        print("Error: Please check operator")
            new_df= df[boolean]
            return(new_df)
        elif movs.loc[x][4] == 'or':
            boolean = (movs.loc[x-1][3])|(movs.loc[x][3])
            if(len(movs.index) > 1):
                for y in range(2, len(movs.index)):
                    if(movs.loc[y][4] == 'and'):
                        boolean = (boolean) & (movs.loc[y][3])
                    elif(movs.loc[y][4] == 'or'):
                        boolean = (boolean) | (movs.loc[y][3])
                    else:
                        print("Error: Please check operator")
            new_df= df[boolean]
            return(new_df)
```
